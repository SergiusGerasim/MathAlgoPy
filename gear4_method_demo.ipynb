```python
X = [0,3]

def dy1dx(x, y1, y2, *args):
    return (-1)*y2 * np.sqrt(y1**2 + y2**2)

def sol_y1(x):
    print(x)
    return np.sqrt(2) * np.cos(x * np.sqrt(2) + (np.pi / 4))


def dy2dx(x, y1, y2, *args):
    return y1 * np.sqrt(y1**2 + y2**2)

def sol_y2(x, *args):
    return np.sqrt(2) * np.sin(np.sqrt(2) * x + (np.pi / 4))


def dydx(x, y1, y2):
    return np.array([dy1dx(x, y1, y2),dy2dx(x, y1, y2)])
SOL = np.array([sol_y1(X[0]), sol_y2(X[0])])
```

```python
hi = 0.05
x, y = Gear_iter(hi,0, N = int(3/hi), dydx=dydx,SOL=SOL,function_count=2)
trace1 = go.Scatter(x=x, y=y[0], name="1")
trace2 = go.Scatter(x=x, y=y[1], name="2")
ex_t1 = go.Scatter(x=x, y = sol_y1(X), name='ex1')
graph([trace1,trace2,ex_t1])
```
